## Introduction 
In this notebook we will be exploring datasets. SNAP is a federal nutrition program that provides monetary aid to supplement grocery bills for low income families in the US. We will be looking at SNAP datasets from 2007,2017, and 2018. Each datasets contains information about each household that recieved benefits within these years. We will be looking at Cherry County in Nebraska, and San Juan County in New Mexico. These Counties are of interst due to previous geographical analysis highlighting these counties as hot or cold spots for SNAP benefits. Cherry County is considered a cold spot due to the decrease in demand for SNAP beneficiaries. San Juan County on the other hand is a hot spot because there are many low income households that need SNAP benefits.
